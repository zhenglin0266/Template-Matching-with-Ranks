```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
start_time = datetime.datetime.now() 
import math

n = 1000
x = [(i+1)/n for i in range(n)]
e = np.random.normal(0, 1, n)
y = [e[i] if (x[i] <=0.25 or x[i]>=0.75) else 4*x[i]-1+e[i] if x[i]<= 0.5 else 3-4*x[i]+e[i] for i in range(n)]
f = [0 if (x[i] <= (0.25) or x[i]>=(0.75)) else 4*(x[i])-1 if x[i]<= (0.5) else 3-4*(x[i]) for i in range(n)]
```

```python
plt.rcParams['figure.figsize'] = (6,8)

plt.subplot(2,2,1)
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('Filter',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,2)
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('Normal noise',fontsize=14)  
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,3)
e = np.random.standard_t(3, size=n)
y = [e[i] if (x[i] <=0.25 or x[i]>=0.75) else 4*x[i]-1+e[i] if x[i]<= 0.5 else 3-4*x[i]+e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('$T_3$ noise',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,4)
e = np.random.standard_cauchy(n)
y = [e[i] if (x[i] <=0.25 or x[i]>=0.75) else 4*x[i]-1+e[i] if x[i]<= 0.5 else 3-4*x[i]+e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.legend(bbox_to_anchor=(1.02, 0), loc=3, borderaxespad=0,fontsize=14)
plt.ylim(-5, 5)
plt.title('Cauchy noise',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show() 
```

```python
f1 = [(1-(4*(x[i]-0.5))**2)**3 for i in range(n)] 
f = [f1[i] if f1[i]>0 else 0 for i in range(n)] 
plt.rcParams['figure.figsize'] = (6,8)

plt.subplot(2,2,1)
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('Filter',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,2)
e = np.random.normal(0, 1, n)
y1 = [(1-(4*(x[i]-0.5))**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('Normal noise',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,3)
e = np.random.standard_t(3, size=n)
y1 = [(1-(4*(x[i]-0.5))**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('$T_3$ noise',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(2,2,4)
e = np.random.standard_cauchy(n)
y1 = [(1-(4*(x[i]-0.5))**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.legend(bbox_to_anchor=(1.02, 0), loc=3, borderaxespad=0,fontsize=14)
plt.ylim(-5, 5)
plt.title('Cauchy noise',fontsize=14) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show() 
```
