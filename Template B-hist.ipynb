```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import scipy.stats as stats
import matplotlib.mlab as mlab 
import math

plt.rcParams['figure.figsize'] = (4, 3)

plt.subplot(1,2,1)
mu = 0
sigma = np.sqrt(1/40)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
score = []
with open("Template B_normal_p.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma)) 
plt.xlabel(r'$\sqrt{n}(\hat{\theta}_n-\theta^*$)',fontsize =11) 
plt.title('Pearson') 

plt.subplot(1,2,2)
mu = 0
sigma = np.sqrt(2.9781/(10.5805**2))
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
score = []
with open("Template B_normal_rank.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma)) 
plt.xlabel(r'$\sqrt{n}(\hat{\theta}_n-\theta^*$)',fontsize =11) 
plt.title('Spearman')  

plt.tight_layout()
plt.show() 
```
