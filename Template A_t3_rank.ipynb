```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
start_time = datetime.datetime.now()

n = 10000
m = int(0.1*n)
err_res = []
for j in range(500):
    x = [(i+1)/n for i in range(n)]
    e = np.random.standard_t(3, size=n)
    y = [e[i] if (x[i] <=0.25 or x[i]>=0.75) else 4*x[i]-1+e[i] if x[i]<= 0.5 else 3-4*x[i]+e[i] for i in range(n)]
    yr = np.searchsorted(np.sort(y), y)
    yr=yr+1
    res = []
    for a in range(-m,m):
        f = [0 if (x[i] <= (0.25+a/n) or x[i]>=(0.75+a/n)) else 4*(x[i]-a/n)-1 if x[i]<= (0.5+a/n) else 3-4*(x[i]-a/n) for i in range(n)]
        inn = np.inner(yr,f)
        res.append(inn)
    max_res = max(res)
    in_res = [i for i in range(2*m) if res[i]==max_res]
    lo_res = random.sample(in_res, 1)[0]
    err = (lo_res-m)/n
    err_res.append(err)
    print(j)

with open("Template A_t3_rank.txt", "w") as f:
    for s in err_res:
        f.write(str(s) +"\n")  

        
# score = []
# with open("file.txt", "r") as f:
#     for line in f:
#         score.append(float(line.strip()))
   
end_time = datetime.datetime.now()    
interval = (end_time-start_time).seconds
final_time = interval/60.0
print ('Time:\t', final_time)   
```
