```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random

plt.rcParams['figure.figsize'] = (13.5, 4.5)

plt.subplot(1,3,1)
score1 = []
with open("Template A_normal_p.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template A_normal_rank.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
        
score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
all_data = [score1, score2]
labels = ["Pearson", "Spearman"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels)  
plt.yscale('log')
plt.title('Normal noise',fontsize=16)
colors = ['pink', 'lightblue']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
    
plt.tick_params(labelsize=14)
plt.xticks(rotation=30)     
plt.ylabel(r'Estimation error $|\hat{\theta}_n-\theta^*|$',fontsize=14) 
plt.yticks(fontsize=14)

plt.subplot(1,3,2)
score1 = []
with open("Template A_t3_p.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template A_t3_rank.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
        
score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
all_data = [score1, score2]
labels = ["Pearson", "Spearman"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels)  
plt.yscale('log')
plt.title('$T_3$ noise',fontsize=16)
colors = ['pink', 'lightblue']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
    
plt.tick_params(labelsize=14)
plt.xticks(rotation=30)     
plt.yticks(fontsize=14)

plt.subplot(1,3,3)
score1 = []
with open("Template A_cauchy_p.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template A_cauchy_rank.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
        
score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
all_data = [score1, score2]
labels = ["Pearson", "Spearman"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels)  
plt.yscale('log')
plt.title('Cauchy noise',fontsize=16)
colors = ['pink', 'lightblue']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
    
plt.tick_params(labelsize=14)
plt.xticks(rotation=30)     
plt.yticks(fontsize=14)
plt.show()
```
